```python
import matplotlib.pyplot as plt

plt.hist(np.random.exponential(1, 100000), density=True, bins=100, histtype="step", color="blue", label="exponential")
plt.hist(np.random.rand(100000), density=True, bins=100, histtype="step", color="green", label="uniform")
plt.hist(np.random.normal(0, 1, 100000), density=True, bins=100, histtype="step", color="red", label="normal")
plt.legend(loc = "upper left")
plt.title("Random distributions")
plt.xlabel("Value")
plt.ylabel("Density")
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D

x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = X**2 + Y**2

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis')

plt.title("3D Plot of Z = X^2 + Y^2")
plt.show()
```

```python
import pandas as pd
import seaborn as sns

# Sample data for demonstration, replace it with the actual data
data = np.random.rand(100, 6)  # Replace with actual data
columns = ['HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed']
df = pd.DataFrame(data, columns=columns)

# Calculate Pearson and Spearman correlations
pearson_corr = df.corr(method='pearson')
spearman_corr = df.corr(method='spearman')

# Visualize with heatmaps
plt.figure(figsize=(14, 5))

plt.subplot(1, 2, 1)
sns.heatmap(pearson_corr, annot=True, cmap='coolwarm', cbar_kws={'label': 'Correlation'})
plt.title('Pearson Correlation Heatmap')

plt.subplot(1, 2, 2)
sns.heatmap(spearman_corr, annot=True, cmap='coolwarm', cbar_kws={'label': 'Correlation'})
plt.title('Spearman Correlation Heatmap')

plt.tight_layout()
plt.show()
```
